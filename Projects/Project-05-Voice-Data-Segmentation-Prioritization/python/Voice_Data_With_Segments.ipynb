```python
import pandas as pd

df = pd.read_csv('/content/Combined with themes - combined_with_themes.csv')
```

```python
def segment_tag(row):

    val1 = str(row['employee_status']).lower()
    val2 = str(row['employee_job_title']).lower()

    if 'regular' in val1 and ('associate' in val2 or 'worker' in val2):
        return 'Full-Time Associate'
    elif 'contract' in val1 and ('packer' in val2 or 'picker' in val2):
        return 'Contract Packer/Picker'
    elif 'regular' in val1 and 'stower' in val2:
        return 'Full-Time Stower'
    elif 'part_time' in val1 and 'stower' in val2:
        return 'Part-Time Stower'
    elif 'regular' in val1 and ('picker' in val2 or 'packer' in val2):
        return 'Full-Time Picker/Packer'
    elif 'part_time' in val1 and ('picker' in val2 or 'packer' in val2):
        return 'Part-Time Picker/Packer'
    elif 'manager' in val2:
        return 'Management'
    else:
        return 'Other'

df['segment'] = df.apply(segment_tag, axis=1)
```

```python
df.to_csv('Voice_Data_with_Segments.csv', index=False)
```
