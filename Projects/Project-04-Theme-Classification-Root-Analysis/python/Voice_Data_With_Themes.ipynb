```python
import pandas as pd
```

```python
df = pd.read_csv("/content/combined_sentiment_dataset.csv")
```

```python
theme_keywords = {
    'Burnout': ['tired', 'exhausted', 'overworked', 'no break', 'stress', 'robot', 'repetitive', 'inhumane', 'long hours', 'demanding',
                'fatigue', 'strain', 'monotonus', 'long'],
    'Infrastructure': ['pressure', 'strict', 'metrics', 'targets', 'rate', 'safety', 'injury', 'numbers', 'unsafe', 'speed',
                              'risks', 'expectations', 'environment', 'tempature'],
    'Management': ['micromanage', 'favoritism', 'toxic', 'miscommunication', 'communication', 'leadership', 'management', 'support',
                   'lack', 'hostile', 'unfair', 'instability'],
    'Benefits': ['good pay', 'great pay', 'pto', 'tuition', 'benefits', 'insurance']
}
```

```python
def tag_themes(text, theme_keywords):
    matched_themes = []
    for theme, keywords in theme_keywords.items():
        if any(keyword in str(text).lower() for keyword in keywords):
            matched_themes.append(theme)
    return ', '.join(matched_themes) if matched_themes else ''
```

```python
df['Theme'] = df['cleaned_text'].apply(lambda x: tag_themes(x, theme_keywords))
```

```python
df.to_csv('combined_with_themes.csv', index=False)
```
